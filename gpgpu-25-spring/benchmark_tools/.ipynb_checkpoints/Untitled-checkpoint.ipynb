```python
# Donnes mmoire brutes (en MiB)
raw_mem = {
    "Version 1": [
        {"Host-to-Device": 374.49, "Device-to-Host": 373.32},
        {"Host-to-Device": 372.00, "Device-to-Host": 370.00},
    ],
    "Version 2": [
        {"Host-to-Device": 370.0, "Device-to-Host": 368.0},
        {"Host-to-Device": 368.0, "Device-to-Host": 367.0},
    ],
    "Version 3": [
        {"Host-to-Device": 365.0, "Device-to-Host": 360.0},
    ],
}

avg_mem = {}
for version, video_list in raw_mem.items():
    df = pd.DataFrame(video_list)
    avg_mem[version] = df.mean()

df_mem_avg = pd.DataFrame(avg_mem).T

# --- Plot transferts mmoire ---
fig, ax = plt.subplots(figsize=(10, 6))
df_mem_avg.plot(kind='bar', ax=ax)

ax.legend(loc='upper left', bbox_to_anchor=(1.02, 1))
plt.xticks(rotation=0)

ax.set_title("Moyenne des transferts mmoire CUDA")
ax.set_ylabel("Taille totale transfre (MiB)")
ax.grid(True, axis='y')
plt.tight_layout()
plt.show()
```
